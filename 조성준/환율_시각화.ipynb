```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x="Date", y="USD/KRW", label="USD/KRW")
sns.lineplot(data=df, x="Date", y="PPI", label="PPI")
sns.lineplot(data=df, x="Date", y="CPI", label="CPI")
sns.lineplot(data=df, x="Date", y="Base Interest Rate", label="KR Interest Rate")
sns.lineplot(data=df, x="Date", y="US Federal Funds Rate", label="US Fed Rate")
plt.title("USD/KRW vs Key Indicator Time Changes")
plt.legend()
plt.tight_layout()
plt.show()
```
