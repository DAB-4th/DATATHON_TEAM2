# 데이터 merge

```python
# 분기별 데이터 합치기
df_q = pd.merge(df1, df_gdp_growth, on='TIME')
df_q
```

```python
df_d = pd.merge(df_kospi, df_3year, on='TIME', how='outer')
df_d = pd.merge(df_d, df_10year, on='TIME', how='outer')
df_d = pd.merge(df_d, df_news, on='TIME', how='outer')
df_d
```

```python
# datetime 객체 만들기
df_d['DATE'] = pd.to_datetime(df_d['TIME'], format='%Y%m%d')
```

```python
# TIME을 datetime으로 객체로 변환
df_q['START_DATE'] = pd.to_datetime(df_q['TIME'].str[:4]+'-'+df_q['TIME'].str[4:].replace({'Q1': '01', 'Q2': '04', 'Q3': '07', 'Q4': '10'}) + '-01')
df_q
```

```python
# 3. GDP 데이터를 일자별로 확장
expanded_q = []

for i in range(len(df_q) - 1):
    start = df_q.loc[i, 'START_DATE']
    end = df_q.loc[i + 1, 'START_DATE'] - pd.Timedelta(days=1)
    dates = pd.date_range(start, end)
    for d in dates:
        expanded_q.append({'DATE': d, 'GDP(십억원)': df_q.loc[i, 'GDP(십억원)'], 'GNI(십억원)':df_q.loc[i, 'GNI(십억원)'], 'GDP_GROWTH(%)':df_q.loc[i, 'GDP_GROWTH(%)']})
```

```python
# 마지막 분기 처리
last_start = df_q.iloc[-1]['START_DATE']
last_dates = pd.date_range(last_start, last_start + pd.offsets.QuarterEnd(0))
for d in last_dates:
    expanded_q.append({'DATE': d, 'GDP(십억원)': df_q.loc[i, 'GDP(십억원)'], 'GNI(십억원)':df_q.loc[i, 'GNI(십억원)'], 'GDP_GROWTH(%)':df_q.loc[i, 'GDP_GROWTH(%)']})
```

```python
df = pd.DataFrame(expanded_q)
df
```

```python
df_d = df_d.drop(columns='TIME')
df_d
```

```python
# 1차 merge
df_merge = pd.merge(df, df_d, on='DATE', how='outer')
df_merge
```

```python
df_eco['DATE'] = pd.to_datetime(df_eco['TIME'], format='%Y%m')
df_eco
```

```python
# 확장용 리스트 만들기
expanded_rows = []

for i in range(len(df_eco)):
    start_date = df_eco.loc[i, 'DATE']
    # 다음 달 1일 - 하루 = 이번 달 말일
    if i < len(df_eco) - 1:
        end_date = df_eco.loc[i + 1, 'DATE'] - pd.Timedelta(days=1)
    else:
        # 마지막 달은 해당 달의 말일까지
        end_date = start_date + pd.offsets.MonthEnd(0)

    # 날짜 생성
    dates = pd.date_range(start=start_date, end=end_date)

    # 같은 값 반복
    for d in dates:
        expanded_rows.append({
            'DATE': d,
            '경제심리지수': df_eco.loc[i, '경제심리지수']
        })

# 최종 데이터프레임
daily_df_eco = pd.DataFrame(expanded_rows)
daily_df_eco
```

```python
# 2차 merge
df_merge = pd.merge(df_merge, daily_df_eco, on='DATE', how='left')
df_merge
```

```python
df_merge.to_csv('merge.csv', encoding='UTF-8', index=False)
```
